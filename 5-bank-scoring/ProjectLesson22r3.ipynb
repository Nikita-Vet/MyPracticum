## Исследование данных

### OHE кодирование

Произведем кодирование категориальных величин методом OHE.

```python
df_ohe = pd.get_dummies(df, drop_first=True)
```

### Разделение на обучающие и целевые признаки

Разделим признаки на обучающие и целевые.

```python
features = df_ohe.drop(['Exited'], axis=1)
target = df_ohe['Exited']
```

### Исследуем баланс классов

```python
# оценим количество ушедщих клиентов от общего их числа
target.mean()
```

```python
_ = target.value_counts(normalize=True).plot(kind='bar', figsize=(5, 3))
```

### Разделениеисходных данных на обучающие, валидационные и тестовые выборки 

Разделим исходные данные на обучающую, валидационную и тестовую выборки в пропорции 0.6, 0.2 и 0.2.

```python
# отделим 40% данных для валидационной и тестовой выборки
features_train, features_sum, target_train, target_sum = train_test_split(
    features, target, test_size=0.4, random_state=12345, stratify=target) 
```

```python
print(features_train.shape)
print(features_sum.shape)
```

```python
# разделим оставшиеся 40% данных поровну на валидационные и тестовые выборки
features_valid, features_test, target_valid, target_test = train_test_split(
    features_sum, target_sum, test_size=0.5, random_state=12345, stratify=target_sum)
```

```python
print(features_valid.shape)
print(features_test.shape)
```

```python
print("Доля миноритарного признака в тренировочной выборке", target_train.mean())
print("Доля миноритарного признака в валидационной выборке", target_valid.mean())
print("Доля миноритарного признака в проверочной выборке", target_test.mean())
```

### Масштабирование данных

Произведем масштабирование обучающих признаков

```python
pd.options.mode.chained_assignment = None  # отключим предупреждение SettingWithCopyWarning
```

```python
scaler = StandardScaler()
scaler.fit(features_train[num_cols])
features_train.loc[:, num_cols] = scaler.transform(features_train[num_cols])
features_valid.loc[:, num_cols] = scaler.transform(features_valid[num_cols])
features_test.loc[:, num_cols] = scaler.transform(features_test[num_cols])
```

```python
# визуализируем тренировочную выборку после кодирования и масштабирования
features_train.sample(10)
```

Предварительно, для получения оценочной метрики используем модель DummyClassifier

```python
model_dc = DummyClassifier(strategy='constant', constant=1, random_state=12345)
model_dc.fit(features_train, target_train)

predicted_train = model_dc.predict(features_train)
accuracy_valid_dc = accuracy_score(target_train, predicted_train)
f1_result_dc = f1_score(target_train, predicted_train)

print("Accuracy модели на тренировочной выборке:", accuracy_valid_dc)
print("F1 модели на тренировочной выборке:", f1_result_dc)
```

### Моделирование несбалансированных данных

#### Дерево решений 

```python
model_dtree = DecisionTreeClassifier(random_state=12345)

params = {'max_depth': range(3, 10),
          'min_samples_leaf': range(1, 10)}

grid_model_dtree = GridSearchCV(model_dtree,
                                params,
                                cv=5,
                                scoring='f1',
                                verbose=1,
                                n_jobs=-1)
```

```python
%%time

grid_model_dtree.fit(features_train, target_train)
```

```python
grid_model_dtree.best_score_
```

```python
grid_model_dtree.best_params_
```

```python
grid_model_dtree.best_estimator_
```

```python
predict_train = grid_model_dtree.best_estimator_.predict(features_train)
print('Точность на тренировочном наборе', accuracy_score(target_train, predict_train))
print('F1 на тренировочном наборе', f1_score(target_train, predict_train)) 
print() 
predict_valid = grid_model_dtree.best_estimator_.predict(features_valid)
print('Точность на валидационном наборе', accuracy_score(target_valid, predict_valid))
print('F1 на валидационном наборе', f1_score(target_valid, predict_valid))
```

```python
cm = confusion_matrix(target_valid, predict_valid)

plt.figure(figsize=(6,3))
sns.heatmap(data=cm, linewidth=.5, annot=True, fmt="g", cmap="Set1")
plt.title("Оценка модели с использованием Confusion Matrix",fontsize=12, pad=12, fontweight="black")
plt.ylabel("Целевые показатели")
plt.xlabel("Прогноз модели")
plt.show()
```

##### Исследуем метрику AUC-ROC

```python
predict_valid_auc = grid_model_dtree.best_estimator_.predict_proba(features_valid)[:,1]


df_actual_predicted = pd.concat([pd.DataFrame(np.array(target_valid), columns=["target_valid"])])

df_actual_predicted.index = target_valid.index

fpr, tpr, thresholds = roc_curve(df_actual_predicted["target_valid"], predict_valid_auc)
auc = roc_auc_score(df_actual_predicted["target_valid"], predict_valid_auc)

plt.figure(figsize=(8, 6))
plt.plot(fpr, tpr, label=f"AUC = {auc:.2f}", color="green")
plt.plot([0, 1], [0, 1], linestyle="--", color="black")
plt.xlabel("Ложно положительная оценка")
plt.ylabel("Истинно положительная оценка или «полнота»")
plt.title("ROC-кривая", pad=20, fontweight="black")
plt.legend()
plt.show()
```

##### Исследуем влияние переменных на модель

### Уменьшение выборки

Зададим функцию для уменьшения доли мажоритарного признака.

```python
def downsample(features, target, fraction):
    features_zeros = features[target == 0]
    features_ones = features[target == 1]
    target_zeros = target[target == 0]
    target_ones = target[target == 1]

    features_downsampled = pd.concat(
        [features_zeros.sample(frac=fraction, random_state=12345)] + [features_ones])
    target_downsampled = pd.concat(
        [target_zeros.sample(frac=fraction, random_state=12345)] + [target_ones])
    
    features_downsampled, target_downsampled = shuffle(
        features_downsampled, target_downsampled, random_state=12345)
    
    return features_downsampled, target_downsampled
```
